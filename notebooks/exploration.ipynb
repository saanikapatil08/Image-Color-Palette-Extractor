# Image Color Palette Extractor - Exploration Notebook

This notebook is used for initial exploration, experimentation, and debugging before building the final Streamlit app.

## Goals
- Load and visualize images
- Convert image to RGB pixel arrays
- Run K-Means clustering manually
- Visualize extracted clusters


```python
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

img = Image.open("/Users/saanika/Saanika_Patil/Git/Image-Color-Palette-Extractor/images.jpg")   #adding image from local system
plt.imshow(img)
plt.axis("off")

pixels = np.array(img).reshape(-1, 3)

kmeans = KMeans(n_clusters=5).fit(pixels)
colors = kmeans.cluster_centers_.astype(int)
colors
```
